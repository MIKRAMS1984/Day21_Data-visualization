# Data Viz

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titani data from seaborn library
df  = sns.load_dataset('titanic')   
df.plot(x = 'age', y = 'fare', kind='scatter', color = 'red')
```

```python
df.plot(x = 'age', y = 'fare', kind='scatter', color = '#0ba334')
```

```python
df.plot(x = 'age', y = 'fare', kind='scatter', color = '#1A73E8')
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['age'].mean()
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('fare')
```

```python
# seaborn plots
sns.barplot(data=df, x='survived', y='fare')
```

```python
sns.barplot(data=df, x='sex', y='fare')
```

```python
sns.barplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.barplot(data=df, x='sex', y='age', hue='class')
```

```python
sns.boxplot(data=df, x='sex', y='age', hue='class')
```

```python
sns.scatterplot(data=df, x='sex', y='age', hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age', hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age', hue='sex')
```

```python
sns.scatterplot(data=df, y='fare', x='age', hue='class')
```
